```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = x*1
y
```

```python
plt.plot(x,y, color='g',linewidth=8)
plt.title('X vs Y')
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.show()
```

```python
#Dot plot
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
c = np.arange(2,20)
c
```

```python
d = 3*c
d
```

```python
plt.plot(c,d,color = "g", linestyle=':', linewidth=5)
plt.title('c vs d')
plt.xlabel('This is c axis')
plt.ylabel('This is d axis')
plt.show()
```

```python
#dual line plot
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
x
```

```python
#Dual lines plot
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title('2 Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
## Pie-Chart
```

```python
fruit = ['Apple','Orange','Mango','Guava']
quantity = [67,47,100,29]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
fruit = ['Apple','Mango','Guava','Pineapple','Orange']
quantity=[10,5,35,21,80]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python

plt.pie(quantity,labels=fruit,autopct='%0.1f%%',
       colors=['yellow','blue','green','red','orange'])
plt.show()
```

```python
# Doughnut Chart
```

```python
fruit = ['Apple','Orange','Mango','Guava','Pineapple']
quantity = [67,34,100,29,54]

plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
