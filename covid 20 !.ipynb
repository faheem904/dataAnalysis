```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
#importing main data set
df = pd.read_csv('covid_19_india.csv',parse_dates=['Date'], dayfirst = True)
```

```python
df.head()
```

```python
#keeping only required columns
df = df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
#renaming column names
df.columns = ['date','state','cured','deaths','confirmed']
```

```python
# Looking at the earlier dates
df.head()
```

```python
#looking at the latest dates
df.tail()
```

```python
#current date 
lastday = df[df.date == '2020-07-17']
```

```python
#today
```

```python
lastday.head()
```

```python
#Sorting data w.r.t number of confirmed cases
max_confirmed_cases=lastday.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
#Getting states with maximum number if confired cases
top_states_confirmed = max_confirmed_cases[0:7]
```

```python
#making bar-plot for states with top confirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='confirmed',data=top_states_confirmed,hue='state')
plt.show()
```

```python
#sorting data w.r.t number of death cases
max_death_cases=lastday.sort_values(by='deaths',ascending=False)
max_death_cases
```

```python
# Getting states with maximum number of death cases
top_states_death=max_death_cases[0:7]
```

```python
# Making bar-plot for states with top death cases
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(x='state',y='deaths',data=top_states_death,hue='state')
plt.show()
```

```python
#sorting data w.r.t number of cured cases
max_cured_cases=lastday.sort_values(by='cured',ascending=False)
max_cured_cases
```

```python
# Getting states with maximum number of cured cases
top_states_cured=max_cured_cases[0:7]
```

```python
# Making bar-plot for stated with top death cases
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(x='state',y='cured',data=top_states_cured,hue='state')
plt.show()
```
