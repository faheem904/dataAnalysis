```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
#import main dataset
df = pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
# Keeping only required columns
df = df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
# renaming column names
df.columns = ['date','state','cured','deaths','confirmed']
```

```python
lastday = df[df.date == '2020-07-17']
```

```python
lastday.head()
```

```python
##Sorting data w.r.t number of comfirmed cases
max_confirmed_cases=lastday.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
# Getting states with maximum number of confirm cases
top_stats_confirmed=max_confirmed_cases[0:5]
top_stats_confirmed
```

```python
# Making bar-plot for states with top confirmed cases
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x='state',y='confirmed',data = top_stats_confirmed, hue='state')
plt.show()
```

```python
# Sorting data w.r.t number of death cases
max_death_cases=lastday.sort_values(by='deaths',ascending=False)
max_death_cases
```

```python
# Getting states with maximum number of death cases
top_states_death=max_death_cases[0:5]
```
